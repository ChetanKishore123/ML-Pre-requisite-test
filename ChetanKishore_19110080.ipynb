# Question 1

Given a matrix **n X k** and a matrix **k X m**.

For multiplication we get result **n X m** matrix, where 1st row of n X m matrix is multiplied to 1st column of 2nd matrix. Hence we need to do k multiplications for it and so on. 

For the complete matrix we need to do **n * k * m** multiplications.

Similarly we need to do **n * (k-1) * m** addition operations.

#Question 2

```python
import numpy as np
import time
import random
n=random.randint(1,10)
k=random.randint(1,10)
m=random.randint(1,10)
M1=np.random.randint(100,size=(n,k))
M2=np.random.randint(100,size=(k,m))
t1=time.time()
ans=np.dot(M1,M2)
t2=time.time()
print(ans)
print("Time using numpy is",t2-t1)
t1=time.time()
for x in range(n):
  for y in range(m):
    ans[x][y]=0
    for z in range(k):
      ans[x][y]+=M1[x][z]*M2[z][y]
t2=time.time()
print(ans)
print("Time using direct multiplication is",t2-t1)
```

#Question 3

Here I am using the method whose complexity is O(n) and takes about n/2 iterations.

```python
n=random.randint(1,20) #length of list
a=[]
for i in range(n):
  a.append(random.randint(1,10000))
b=a.copy()
print(b)
t1=time.time()
if n%2 == 0:
  m1=m2=max(b)
  b.remove(m2)
  for i in range(1,int(n/2)+1):
    m1=m2
    m2=max(b)
    b.remove(m2)
  ans=(m1+m2)/2
  print(ans)
else:
  for i in range(int((n+1)/2)):
    m=max(b)
    b.remove(m)
  ans=m
  print(ans)
t2=time.time()
print("Time taken by brute force method is",t2-t1)
t1=time.time()
ans=np.median(a)
print(ans)
t2=time.time()
print("Time taken by numpy method is",t2-t1)
```

#Question 4

[2xy + $y^3$$cosx$ , $x^2$ + 3$y^2$$sinx$] is the gradient of the function $x^2$y + $y^3$$sinx$ 

#Question 5


```python
import jax
import jax.numpy as npj
def f(x,y):
  return ((x**2)*y)+((y**3)*(npj.sin(x)))
def xdev(x,y):
  return 2*x*y+(y**3)*(npj.cos(x))
def ydev(x,y):
  return (x**2)+npj.sin(x)*3*(y**2)    
dif_x=jax.grad(f,0)
x=random.random()
y=random.random()
# print(x,y)
print(" gradient from jax method is",dif_x(x,y))
print("Gradient by our method is",xdev(x,y))
dif_y=jax.grad(f,1)
print(" gradient from jax method is",dif_y(x,y))
print("Gradient by our method is",ydev(x,y))
```

#Question 12

Broadcasting: 
The matrix whose dimension is lower, it's values are broadcast to make the basic matrix operations compatible.


```python
x=np.array([[1,2,3],[3,1,0],[2,4,1]])#a 3X3 matrix 
y=np.array([2,0,1])#a 1X3 matrix
z=x+y
print(z)
'''Here we get the results as if the y array has been broadcast to form 
[[2 0 1]
 [2 0 1]
 [2 0 1]]'''
z=x*y
print(z)
```
